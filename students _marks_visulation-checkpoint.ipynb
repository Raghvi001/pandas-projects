```python
import pandas as pd 
df =pd.read_csv("students.csv")
df
```

```python
df["avg"]=df[[" maths","english","science"]].mean(axis=1)
```

```python
import matplotlib.pyplot as plt 
plt.bar(df["name"],df["avg"])
plt.xlabel("students")
plt.ylabel("average marks")
plt.title("average marks per student ")
plt.show()
```

```python
subject_avg=df[[" maths","english","science"]].mean(axis=0)
import matplotlib.pyplot as plt 
plt.bar(subject_avg.index,subject_avg.values)
plt.xlabel("subjects")
plt.ylabel("average marks ")
plt.title("average marks by the subject")
plt.show()
```

```python
df["marks"]=df[[" maths","english","science"]].sum(axis=1)
```

```python
df["result"]=df["marks"].apply(lambda x :
                               "pass" if x>75 else "fail")
                               
```

```python
df["result"]
```

```python
result_counts=df["result"].value_counts()
import matplotlib.pyplot as plt 
plt.pie(
    result_counts.values,
labels=result_counts.index,
autopct="%1.1f%%",
startangle=90)
plt.title("pass vs fail distribution")
plt.show()
```
