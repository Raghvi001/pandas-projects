```python
import pandas as pd
df=pd.read_csv("students.csv")
subject_avg=df[[" maths","english","science"]].mean()
import matplotlib.pyplot as plt
plt.plot(
    subject_avg.index,
    subject_avg.values,
    marker="o"
)
plt.xlabel=("subject")
plt.ylabel=("average marks")
plt.title=("average marks trend across subjects")
plt.grid(True,linestyle="--",alpha=0.5)
plt.show()
            

```

```python
import matplotlib.pyplot as plt
plt.scatter(df[" maths"],df["science"])
plt.xlabel=(" maths marks")
plt.ylabel=("science marks")
plt.title=("maths vs physics marks")
plt.show()
```

```python
plt.scatter(df[" maths"],df["science"],alpha=0.7)
plt.grid(True)
```

```python
plt.scatter(df[" maths"],df["english"],c=df[" maths"],cmap="viridis")
```

```python
import matplotlib.pyplot as plt 
plt.scatter(
    df[" maths"],
    df["science"],
    c=df[" maths"],
    cmap="viridis",
    alpha=0.7,
)
plt.xlabel=(" maths marks")
plt.ylabel=("science marks")
plt.title=("maths vs escience marks ")
plt.colorbar(label="maths marks")
plt.grid=(True)
plt.show()
    
```
