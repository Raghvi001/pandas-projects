```python
import pandas as pd
df=pd.read_csv("students.csv")
df
```

```python
marks_df=df[[" maths","english","science"]]
corr_matrix=marks_df.corr()
corr_matrix

```

```python
import matplotlib.pyplot as plt
plt.imshow(corr_matrix)
plt.colorbar()
plt.xticks(range(len(corr_matrix)),corr_matrix.columns)
plt.yticks(range(len(corr_matrix)),corr_matrix.columns)
plt.title("correlation")
plt.show()
```

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
df["average_marks"]=df[["maths","english","science"]].mean(axis=1)
df["average_marks"]
```

```python
df["grade"]=df["average_marks"].apply(
    lambda x: "A" if x>90 else
    "B" if x>80 else
    "C" if x>70 else
    "fail")
df["grade"]
```

```python
df["grade"].value_counts()
```

```python
df["grade"].value_counts().plot(kind="bar")
plt.title("grade distribution")
plt.show()
```

```python
subject_average_result=df.groupby("grade")[["maths","science","english"]].mean()
subject_average_result
```

```python
df.columns
```

```python
subject_average_result.plot(kind="bar")
plt.title("average subject marks")
plt.xlabel("result")
plt.ylabel("grade")

plt.show()
```
